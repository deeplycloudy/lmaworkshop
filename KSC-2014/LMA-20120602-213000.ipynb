````python
import matplotlib
%matplotlib qt4
# import matplotlib
# matplotlib.use('nbagg')
#import matplotlib.pyplot as plt
from brawl4d.brawl4d import B4D_startup, redraw
import os
````

````python
data_path = '/data/DC3/flash_sort/'
lma_file = os.path.join(data_path, 'h5_files/2012/Jun/02/LYLOUT_120602_213001_0600.dat.flash.h5')
````

In the cell below, note that the basedate has been set to match the dataset we specified above.

If you are not using data from the WTLMA, then you'll also need to pass <code>ctr_lon=value</code> and <code>ctr_lat=value</code> to B4D_startup.

````python
from datetime import datetime
panels = B4D_startup(basedate=datetime(2012,6,2), ctr_lat=40.4463980, ctr_lon=-104.6368130)
````

````python
import matplotlib.pyplot as plt; plt.show()
````

````python
from brawl4d.LMA.controller import LMAController
lma_ctrl = LMAController()
d, post_filter_brancher, scatter_ctrl, charge_lasso = lma_ctrl.load_hdf5_to_panels(panels, lma_file)
````

### 2131:38-54

````python
h,m,s,dt = 21, 31, 38, 16
panels.panels['tz'].axis((h*3600 + m*60 + s, h*3600 + m*60 + s+dt, 1, 18))
panels.panels['xy'].axis((27, 47, 45, 65))
````

Large flash on north end shows very nicely the rearward tilt of both positive and negative charge.

Positive charge layer slopes from (30.8, 5.0) to (37.4, 7.8). With U = 10 m/s eastward storm motion, dz/dx * U = dz/T is the fall speed. This gives 4 m/s, which is a reasonable value for graupel. However the negative charge, which is presumably carried on ice crystals of negligible fall speed, shows a similar slope, which would not be expected if slope were purely due to sedimentation.


````python
print "Fall speed estimate is U * dz/dx = ", 10 * (7.8 - 5.0) / (37.4 - 30.8)
````

Flashes between 46 and 49 seconds are each small in size and negative over positive where they can be analyzed.

Two flashes at 51-52 seconds again are both -ICs, but at very different altitudes in the storm. Higher negative charge is in the region with the small flashes.

So, the region at 38, 55 is likely the updraft, and then sedimentation of positive-bearing precipitation around this.

This period of the storm begins the onset of decay, where sedimentation and downdrafts take over for updrafts.
