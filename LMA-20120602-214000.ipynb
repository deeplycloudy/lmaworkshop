===============================================================

This notebook demonstrates how to display data included in the lmatools repository.

Then, edit the data_path in the second cell to include `/path/to/output/files/` as defined above. Run all the cells prior to the "Charge Analysis" and try interacting with the plot. You should see some data.


````python
import matplotlib
%matplotlib qt4
# import matplotlib
# matplotlib.use('nbagg')
#import matplotlib.pyplot as plt
from brawl4d.brawl4d import B4D_startup, redraw
import os
````

````python
data_path = '/data/DC3/flash_sort/'
lma_file = os.path.join(data_path, 'h5_files/2012/Jun/02/LYLOUT_120602_214001_0600.dat.flash.h5')
````

In the cell below, note that the basedate has been set to match the dataset we specified above.

If you are not using data from the WTLMA, then you'll also need to pass <code>ctr_lon=value</code> and <code>ctr_lat=value</code> to B4D_startup.

````python
from datetime import datetime
panels = B4D_startup(basedate=datetime(2012,6,2), ctr_lat=40.4463980, ctr_lon=-104.6368130)
````

````python
import matplotlib.pyplot as plt; plt.show()
````

````python
from brawl4d.LMA.controller import LMAController
lma_ctrl = LMAController()
d, post_filter_brancher, scatter_ctrl, charge_lasso = lma_ctrl.load_hdf5_to_panels(panels, lma_file)
````

### Zoom to overview of first minute

````python
h,m,s,dt = 21, 40, 0, 60
panels.panels['tz'].axis((h*3600 + m*60 + s, h*3600 + m*60 + s+dt, 1, 18))
panels.panels['xy'].axis((-150, 150, -150, 150))
````

````python
from brawl4d.LMA.widgets import LMAwidgetController
from IPython.display import display
from brawl4d.LMA.controller import LMAController

lma_tools = LMAwidgetController(panels, lma_ctrl, scatter_ctrl, charge_lasso, d)
display(lma_tools.tools_popup)
````

### Hereford storm

````python
h,m,s,dt = 21, 40, 0, 60*2
panels.panels['tz'].axis((h*3600 + m*60 + s, h*3600 + m*60 + s+dt, 1, 15))
panels.panels['xy'].axis((30, 50, 45, 65))
````

By this point, all flashes are large, and occurr every 5-10 s.

````python
h,m,s,dt = 21, 42, 0, 60*2
panels.panels['tz'].axis((h*3600 + m*60 + s, h*3600 + m*60 + s+dt, 1, 15))
panels.panels['xy'].axis((30, 50, 45, 65))
````

Two upper level flashes just after 42:40 and 43:40. These are +IC flashes, with upward negative leaders.

````python
h,m,s,dt = 21, 44, 0, 60*2
panels.panels['tz'].axis((h*3600 + m*60 + s, h*3600 + m*60 + s+dt, 1, 15))
panels.panels['xy'].axis((34, 54, 45, 65))
````

One last upper flash at 2144:45.

Also now starting to get isolated flashes out of the forward part of the storm. Great example at 2145:37.9, which the NLDN saw as a -21 kA stroke. Note the negative charge is lowered from between 8 and 10 km.

````python
h,m,s,dt = 21, 46, 0, 60*2
panels.panels['tz'].axis((h*3600 + m*60 + s, h*3600 + m*60 + s+dt, 1, 15))
panels.panels['xy'].axis((35, 55, 45, 65))
````

````python
h,m,s,dt = 21, 48, 0, 60*2
panels.panels['tz'].axis((h*3600 + m*60 + s, h*3600 + m*60 + s+dt, 1, 15))
panels.panels['xy'].axis((35, 55, 45, 65))
````

Flash at 2149:34 remains generally elevated, with positive charge near the melting level, but after eastward extending positive leaders there is an attempted CG. NLDN has a negative cloud pulse at about the right time, but doesn't indicate a ground strike.
